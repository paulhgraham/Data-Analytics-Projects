```python
import pandas as pd

df = pd.read_csv('raw_data/employee_data1.csv')
ef = pd.read_csv('raw_data/employee_data2.csv')


frames = [df, ef]
combined_data = pd.concat(frames)

#date of birth format
combined_data['DOB'] = pd.to_datetime(combined_data.DOB)
combined_data['DOB1'] = combined_data['DOB'].dt.strftime('%m/%d/%Y')
combined_data['DOB'] = combined_data['DOB1']


#split name
combined_data[['Name','Last Name']] = combined_data.Name.str.split(' ',expand=True)
combined_data.rename(index=str, inplace = True, columns={'Name': 'First Name'})

#combined_data('First') = combined_data.Name.map(lambda x: x.split()[0])
#combined_data('Last') = combined_data.Name.map(lambda x: x.split()[1])
#combined_data.head()
#name_split = combined['Name'].str(split('',0,true))
#new_column = pd.combined(Name.str.split(' ',None,true),
#name_split = ['First Name','Last Name'])

#SSN format
def ssn_private(row):
    last4 = row[-4:-1]
    return("***-**-" + last4)
combined_data['Masked SSN'] = combined_data['SSN'].map(ssn_private)
combined_data['SSN'] = combined_data['Masked SSN']

#state abb format
combined_data['StateAbb'] = combined_data.State.map(us_state_abbrev)
combined_data['State'] = combined_data['StateAbb']

#order columns
combined_data = combined_data.reindex(['Emp ID', 'First Name', 'Last Name', 'DOB', 'Masked SSN', 'State'], axis=1)

#Drop data
#combined_data.drop(['DOB1','StateAbb','SSN'], axis=1)################### why won't it delete 


combined_data
```
